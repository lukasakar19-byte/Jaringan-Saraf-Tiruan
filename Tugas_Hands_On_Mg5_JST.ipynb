```python
from collections import deque

def bfs(graph, start_node, target_node):
    """Mengimplementasikan algoritma Breadth-First Search (BFS)."""
    # Langkah 1: Inisialisasi Antrean dan Set Visited
    queue = deque([start_node]) # Antrean untuk node yang akan dikunjungi
    visited = {start_node}   # Set untuk melacak node yang sudah dikunjungi
    paths = {start_node: [start_node]} # Dictionary untuk menyimpan jalur dari start_node

    print(f"Memulai BFS dari node: {start_node}")

    while queue:
        current_node = queue.popleft() # Ambil node pertama dari antrean
        print(f"Mengunjungi node: {current_node}")

        # Langkah 2: Periksa Tujuan
        if current_node == target_node:
            print(f"\nNode tujuan ({target_node}) ditemukan!")
            return paths[current_node] # Kembalikan jalur ke node tujuan

        # Langkah 3: Jelajahi Tetangga
        for neighbor in graph.get(current_node, []):
            if neighbor not in visited:
                visited.add(neighbor) # Tandai tetangga sebagai sudah dikunjungi
                queue.append(neighbor) # Tambahkan tetangga ke antrean
                # Perbarui jalur untuk tetangga
                paths[neighbor] = paths[current_node] + [neighbor]
                print(f"  Menambahkan jalan {neighbor} ke antrean dan jalur: {paths[neighbor]}")

    print(f"\nNode tujuan ({target_node}) tidak dapat dijangkau dari {start_node}.")
    return None # Jika antrean kosong dan tujuan tidak ditemukan

# Contoh Penggunaan:
# Definisikan graf sebagai dictionary adjacency list
graph = {
    "Rumah Lukas": ["G.Obos","RTA"],
    "G.Obos": ["Soekarno"],
    "RTA": ["Bundaran Burung"],
    "Soekarno": ["Bundaran Burung"],
    "Bundaran Burung": ["Duta Mall"],
    "Duta Mall": []
}

print("=== Contoh BFS untuk mencari jalur dari Rumah Lukas ke Duta Mall ===")
path_to_duta_mall = bfs(graph, 'Rumah Lukas', 'Duta Mall')
if path_to_duta_mall:
    print(f"Jalur terpendek dari Rumah Lukas ke Duta Mall: {path_to_duta_mall}")
```
