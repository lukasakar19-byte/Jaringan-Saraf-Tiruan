```python
# SMART STUDENT AI - Hari Deadline

# BAGIAN 1 : AGENT
class StudentAgent:
    def __init__(self, name):
        self.name = name

    # Agent menerima kondisi lingkungan
    def perceive(self, env):
        return env

    # Agent mengambil keputusan
    def act(self, perception):
        decision = decision_tree(perception)
        return decision

# BAGIAN 2 : LOGIKA PROPOSISI
def logika_proposisi(ngantuk, lapar):
    # Jika ngantuk DAN lapar -> tidak fokus
    return ngantuk and lapar

# BAGIAN 3 : POHON KEPUTUSAN
def decision_tree(data):
    deadline = data["deadline"]
    kuota = data["kuota"]
    teman = data["teman"]
    ngantuk = data["ngantuk"]
    lapar = data["lapar"]

    tidak_fokus = logika_proposisi(ngantuk, lapar)

    # Struktur pohon keputusan
    if deadline == "dekat":
        if kuota:
            if teman and tidak_fokus:
                return "Nongkrong"
            elif tidak_fokus:
                return "Tidur"
            else:
                return "Mengerjakan Tugas"
        else:
            return "Cari WiFi"
    else:
        if teman:
            return "Nongkrong"
        else:
            return "Santai"


# BAGIAN 4 : JARINGAN SEMANTIK MINI
semantic_network = {
    "Mahasiswa": ["Tugas", "Deadline", "Teman", "Kuota"],
    "Tugas": ["Stres"],
    "Teman": ["Nongkrong"],
    "Deadline": ["Panik"]
}


# BAGIAN 5 : SISTEM PAKAR SEDERHANA
def expert_comment(decision):
    comments = {
        "Nongkrong": "Deadline dekat oi, cemas kau dek.",
        "Tidur": "bobo dulu, tugas bisa nanti (katanya).",
        "Mengerjakan Tugas": "MANTAP, Ini baru teladan.",
        "Cari WiFi": "Cari Wifi dulu bang.",
        "Santai": "Deadline masih lama, santai dulu gak sih."
    }
    return comments.get(decision, "Keputusan tidak diketahui.")


# INPUT LINGKUNGAN
def input_environment():
    print("=== KONDISI MAHASISWA ===")
    ngantuk = input("Ngantuk? (ya/tidak): ").lower() == "ya"
    lapar = input("Lapar? (ya/tidak): ").lower() == "ya"
    deadline = input("Deadline (dekat/lama): ").lower()
    kuota = input("Punya kuota? (ya/tidak): ").lower() == "ya"
    teman = input("Teman ngajak nongkrong? (ya/tidak): ").lower() == "ya"

    return {
        "ngantuk": ngantuk,
        "lapar": lapar,
        "deadline": deadline,
        "kuota": kuota,
        "teman": teman
    }

# PROGRAM UTAMA
def main():
    env = input_environment()

    student_ai = StudentAgent("Lukas")
    perception = student_ai.perceive(env)
    decision = student_ai.act(perception)

    print("\n=== HASIL KEPUTUSAN AI ===")
    print("Keputusan Mahasiswa:", decision)
    print("Komentar Sistem Pakar:", expert_comment(decision))

main()
```
