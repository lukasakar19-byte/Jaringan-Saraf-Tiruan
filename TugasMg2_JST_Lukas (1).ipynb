## Generate Data

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Tambah data ke-21
X1 = np.random.normal(loc=1, scale=10, size=(1000, 1))
print("Nilai X1 : ",X1)

# Generate a synthetic classification dataset
X2, y = make_classification(n_samples=1000, n_features=21, n_informative=10, n_redundant=5, n_classes=2, random_state=42)

#Gabungkan data
X = np.hstack((X1, X2))

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Initialize the Sequential model with an Input layer
model = Sequential([
    Input(shape=(X_train.shape[1],)) # Explicitly define input shape
])

# Add the first hidden layer
model.add(Dense(units=64, activation='relu'))

# Add a second hidden layer
model.add(Dense(units=32, activation='relu'))

# Add the output layer (1 unit with sigmoid activation for binary classification)
model.add(Dense(units=1, activation='sigmoid'))

# Print the model summary to display the architecture
model.summary()
```

Arsitektur jaringan saraf tiruan terdiri dari 22 neuron input, dua hidden layer masing-masing berisi 64 dan 32 neuron dengan fungsi aktivasi ReLU, serta satu neuron output dengan fungsi aktivasi sigmoid untuk klasifikasi biner. Total parameter yang dipelajari model berjumlah 3.585 yang merupakan bobot dan bias dari seluruh koneksi antar neuron. Struktur ini memungkinkan model melakukan ekstraksi fitur bertahap sebelum menentukan hasil klasifikasi.
